# Practical Review + Multiclass Wrap Up

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math
from scipy.optimize import minimize
```

0) Read data file

```python
class_map = {'Iris-setosa\n':0, 'Iris-versicolor\n':1, 'Iris-virginica\n':2}
features = []
classes = []
feature_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
with open('iris.data') as f:
    for line in f:
        s = line.split(',')
        if len(s)==5:
            feature = [float(x) for x in s[:4]]
            features.append(feature)
            classes.append(class_map[s[-1]])
ndata = len(features)
features = np.array(features)
classes = np.array(classes)
padded_features = np.lib.pad(features,((0,0),(1,0)),'constant', constant_values=1)
print padded_features
#print features
#print classes
```

1) Make scatter plot of every single features.

```python
nfeature = len(feature_names)
fig, axarr = plt.subplots(nfeature, nfeature,figsize=(12,12), sharex="col", sharey="row")
for fx in range(nfeature):
    for fy in range(nfeature):
        for iclass in [0,1,2]:
            ax = axarr[fy, fx]
            rows = classes==iclass
            xs = features[rows,fx]
            ys = features[rows,fy]
            ax.plot(xs, ys, '.')
            ax.set_xlabel(feature_names[fx])
            ax.set_ylabel(feature_names[fy])
```

2) Then train each classifier

```python
class Logistic:
    def __init__(self):
        self.w = None
        
    def train(self, features, classes):
        def cost(w):
            s = classes*np.dot(features, w)
            s = np.clip(s,-200,200)
            t = np.log1p(1+np.exp(-s))
            return np.sum(t)
        res = minimize(cost, [0]*len(features[0]))
        self.w = res.x

    
    def classify(self, features):
        s = np.dot(features, self.w)
        s = np.clip(s,-200,200)
        return 1./(1.+math.exp(-s))
```

```python
#Thre are 3 of them we need to train. (2^(3-1)-1)
classifiers = []
for i in range(3):
    #first transform class
    #classes for this classifier
    new_class = classes.copy()
    new_class[classes==i] = 1
    new_class[classes!=i] = -1
    L = Logistic()
    L.train(padded_features, new_class)
    classifiers.append(L)
```

3) Get output from each classifier

```python
outputs = np.zeros((ndata, 3))
for idata, data in enumerate(padded_features):
    for iclassifier, classifier in enumerate(classifiers):
        outputs[idata, iclassifier] = classifiers[iclassifier].classify(data)
        
print outputs
```

4) now compare each one with the template answer

```python
templates = [[1,0,0], [0,1,0], [0,0,1]]
toutput = np.empty((ndata, 3))
for ioutput, output in enumerate(outputs):
    for itemplate, template in enumerate(templates):
        toutput[ioutput, itemplate] = np.sum((output-template)**2)
print toutput
```

5) For each one select the one with minimum index

```python
guesses = np.argmin(toutput, axis=1)
print guesses
```

```python
nfeature = len(feature_names)
fig, axarr = plt.subplots(nfeature, nfeature,figsize=(12,12), sharex="col", sharey="row")
for fx in range(nfeature):
    for fy in range(nfeature):
        for iclass in [0,1,2]:
            ax = axarr[fy, fx]
            rows = classes==iclass
            xs = features[rows,fx]
            ys = features[rows,fy]
            ax.plot(xs, ys, '.')
            ax.set_xlabel(feature_names[fx])
            ax.set_ylabel(feature_names[fy])
```
