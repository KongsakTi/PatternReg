```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math
from scipy.optimize import minimize
```

0) Read data file

```python
class_map = {'Iris-setosa\n':0, 'Iris-versicolor\n':1, 'Iris-virginica\n':2}
features = []
classes = []
feature_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
with open('iris.data') as f:
    for line in f:
        s = line.split(',')
        if len(s)==5:
            feature = [float(x) for x in s[:4]]
            features.append(feature)
            classes.append(class_map[s[-1]])
ndata = len(features)
features = np.array(features)
classes = np.array(classes)
padded_features = np.lib.pad(features,((0,0),(1,0)),'constant', constant_values=1)
print padded_features
#print features
#print classes
```

1) Make scatter plot of every single features.

```python
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# row and column sharing
f, axarr = plt.subplots(4, 4, figsize=(15,15))
nfeatures = len(features[0])
nclass = len(classes)
for ix in xrange(nfeatures):
    for iy in xrange(nfeatures):
        if (ix != iy):
            for iclass in xrange(nclass):
                idx = classes == iclass
                fx = features[idx, ix]
                fy = features[idx, iy]
                axarr[ix, iy].plot(fx, fy, '.')
        if (ix == 0):
            axarr[iy, ix].set_ylabel(feature_names[iy])
        if (iy == nfeatures - 1):
            axarr[iy, ix].set_xlabel(feature_names[ix])
```

2) Then train each classifier

```python
def make_classes(class_spec, org_classes):
    return np.array([class_spec[x] for x in org_classes])
```

```python
class LogisticRegression:
    def __init__(self):
        self.w = None
        
    def train(self, features, classes):
        def cost(w):
            s = np.dot(features, w)
            s = np.clip(s, -200, 200)
            return np.sum(np.log1p(np.exp(-classes*s)))
        self.w = minimize(cost, [0.]*5).x

    def classify(self, features):
        s = np.dot(features, self.w)
        s = np.clip(s, -200, 200)
        return 1./(1 + np.exp(-s))
    
    def classifypm(self, features):
        return (self.classify(features) - 0.5)*2
```

```python
template = np.array([
    [1, -1, -1],
    [-1, 1, -1],
    [-1, -1, 1]
])
# first idx is a class
# second idx is a classifier

nClasses, nClassifier = template.shape

LRs = []
for iClassifier in xrange(nClassifier):
    LR = LogisticRegression()
    LR.train(padded_features, make_classes(template[iClassifier], classes))
    LRs.append(LR)

```

3) Get output from each classifier

```python
def getData(features):
    return [ [LR.classifypm(feature) for LR in LRs] for feature in features ]

datas = getData(padded_features)

# from IPython.display import display
scores = [[np.sum((template[iClass] - data)**2) for iClass in xrange(nClasses)] for data in datas]
cs = np.array(np.argmin(scores, axis = 1))



print cs
print classes


print np.sum(cs == classes)/float(len(cs))

```
