```python
%matplotlib inline
from matplotlib import pyplot as plt
import pydot
from IPython.display import Image, display
```

```python
graph = pydot.Dot(graph_type='digraph')
root = pydot.Node('root')
graph.add_node(root)
```

```python
for i in range(3):
    child = pydot.Node('c%d'%i, )
    graph.add_node(child)
    graph.add_edge(pydot.Edge(root, child))
    for j in range(3):
        grand_child = pydot.Node('gc_%d_%d'%(i,j), label='gc%d'%j )
        graph.add_node(grand_child)
        graph.add_edge(pydot.Edge(child, grand_child))
```
