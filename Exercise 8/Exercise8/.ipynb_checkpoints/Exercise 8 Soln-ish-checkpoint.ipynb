```python
%matplotlib inline
from scipy.io import loadmat
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
```

```python
np.random.seed(12345)
nd = 100
x1p = np.random.randn(nd)+1
x1n = np.random.randn(nd)+2

x2p = x1p *2 + 1 + np.random.randn(nd)
x2n = x1n *2 + 1 + np.random.randn(nd)

x1 = np.concatenate([x1p, x1n])
x2 = np.concatenate([x2p, x2n])

xs = np.vstack([x1, x2]) #now the index for data point is the second index.
ys = np.concatenate([[1]*nd, [-1]*nd])
plt.plot(x1[ys>0], x2[ys>0], '.b')
plt.plot(x1[ys<0], x2[ys<0], '.r')
plt.grid(True)
```

1) Normalize the data so that the mean of each feature is 0 and standard deviation of each feature is 1.

```python
norm_xs = xs.copy()

s = xs[0,:]
s = s - np.average(s)
s = s / np.std(s)

norm_xs[0, :] = s

s = xs[1,:]
s = s - np.average(s)
s = s / np.std(s)

norm_xs[1, :] = s
def plot_data():
    plt.plot(norm_xs[0, :], norm_xs[1, :], '.')
plot_data()
```

2) What is the criteria for picking better axis?

3) What is the size of the covariance matrix? Why?

```python
nfeature, ndata = norm_xs.shape
cov = np.zeros((nfeature, nfeature))
for i in range(nfeature):
    for j in range(i, nfeature):
        cov[i,j] = np.sum(norm_xs[i]*norm_xs[j])/ndata
        cov[j,i] = cov[i,j]
        
print cov
```

4) Find eigenvalue and eigenvector of the covariance matrix.

```python
np.linalg.eigh(cov)
```

5) Draw the eigenvectors on the plot.

```python
t = np.linspace(-5, 5)
e2, e1 = np.linalg.eigh(cov)[1]
xx = t * e1[0]
yy = t * e1[1]
plot_data()
plt.plot(xx,yy, 'g')
xx = t * e2[0]
yy = t * e2[1]
plt.plot(xx,yy, 'r')
plt.axes().set_aspect('equal', 'datalim')
plt.grid(True)
print e1, e2
```

6) Use eigenvector and compute the transformed feature vector.

```python
trans = np.linalg.eigh(cov)[1]
t_xs = np.dot(trans, norm_xs)

plt.plot(t_xs[0, :], t_xs[1, :], '.')
```

```python

plt.plot(t_xs[0, ys>0], t_xs[1, ys>0], '.b')
plt.plot(t_xs[0, ys<0], t_xs[1, ys<0], '.r')
```

```python
#train_labels is in a funny format too
#here is how to use it
labels = D['train_labels'].T
#this is how they represent 6
print labels[0] 
print labels[0,6]

def make_class(c):
    return list(c).index(1)
labels = np.array([make_class(lb) for lb in labels])
print labels
```

```python
images = D['train_patterns']
def show(img):
    im = img.reshape(16,16) #you can try print it
    plt.imshow(im, interpolation='nearest', cmap=cm.Greys)
    
show(images[:,0])
plt.figure()
show(images[:,1])
```

```python
print len(np.average(images, axis=1))
```

```python
images_std = np.std(images, axis=1)
images_mean = np.average(images, axis=1)
norm_images = images - images_mean[:,  np.newaxis]
norm_images = norm_images/images_std[:,  np.newaxis]
```

```python
nfeature, nimages = images.shape
cov = np.dot(images, images.T)/nimages
```

```python
evalue, evec = np.linalg.eigh(cov)
```

```python
for i in range(20):
    plt.figure()
    show(evec[-i] * images_std + images_mean)
```

```python
new_features = np.dot(evec, norm_images)
print new_features.shape
```

```python
reduced_feature = new_features[-50:,:]
```

```python
from sklearn import svm
clf = svm.SVC(decision_function_shape='ovo')
clf.fit(reduced_feature.T, labels)
```

```python
good = 0
bad = 0
for i in range(nimages):
    predict = clf.predict(reduced_feature[:,i].reshape(1, -1))[0]
    if(predict==labels[i]):
        good += 1
    else:
        bad += 1
print good,bad, float(good)/(good+bad)
```
